```python
from sage.all import *

# Precision used for each component in bits
PREC = 52
D = 54
RF = RealField(3068)
ZR = RealField(PREC)

def double_to_hex(f):
    return "0x" + format(struct.unpack('<Q', struct.pack('<d', f))[0], '016x')

def round_nearest(x, bits):
    R = RealField(bits)
    return R(x)

def to_hex_scaled(x):
    R = RealField(PREC)
    scaled = R(2)**7 * x
    return double_to_hex(scaled)

for i in range(128):
    if i < 3:
        pi_inv = RF(0.25) + RF(2) ** (16 * (i - 3)) / RF.pi()
    else:
        pi_inv = RF(2) ** (16 * (i - 3)) / RF.pi()
    
    pn = pi_inv.round()
    pi_frac = pi_inv - pn

    a = round_nearest(pi_frac, PREC)
    b = round_nearest(pi_frac - RF(a), PREC)
    c = round_nearest(pi_frac - RF(a) - RF(b), PREC)
    d = round_nearest(pi_frac - RF(a) - RF(b) - RF(c), PREC)

    # Output as hex
    print("(", to_hex_scaled(a), ",", to_hex_scaled(b), ",", to_hex_scaled(c), ",", to_hex_scaled(d), "),")
```
